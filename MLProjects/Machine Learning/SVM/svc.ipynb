   ## RBF Kernel

The kernel is used to convert a low-dimensional input space into a higher-dimensional space. This is done by mapping the data into a new feature space. In this space, the data will be linearly separable. 

The RBF kernel allows SVMs to implicitly map the input data into a high-dimensional space, where it becomes linearly separable. By applying the RBF kernel, SVMs can find a hyperplane that maximally separates the classes in this transformed space.

The mathematical form of the RBF kernel is:

K(x, y) = exp(-gamma * ||x - y||^2)
