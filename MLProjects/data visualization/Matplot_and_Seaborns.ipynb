```python
#Seaborn

import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
#coreleation

#finds the inter relation between the two variables

df.corr()
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr())
```

```python
#joint plot
# can do univariate analysis and bi variate

sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
#Pair plot
#when we have more features

sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['size'].value_counts()
```

```python
#dist plots

sns.distplot(df['tip'],kde=False,bins=15)
```

```python
#Categorival plots

#boxplot
#violin plot
#count plot
#bar plot
```

```python
sns.countplot(y='smoker',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot('sex','total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot('day','total_bill',data=df,hue='smoker')
```

```python
sns.violinplot('total_bill','day',data=df)
```
